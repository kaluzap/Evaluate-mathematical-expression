Instructions

Given a mathematical expression as a string you must return the result as a number.

Numbers

Number may be both whole numbers and/or decimal numbers. The same goes for the returned result.

Operators

You need to support the following mathematical operators:

    Multiplication *
    Division / (as true division)
    Addition +
    Subtraction -

Operators are always evaluated from left-to-right, and * and / must be evaluated before + and -.

Parentheses

You need to support multiple levels of nested parentheses, ex. (2 / (2 + 3.33) * 4) - -6

Whitespace

There may or may not be whitespace between numbers and operators.

An addition to this rule is that the minus sign (-) used for negating numbers and parentheses will never be separated by whitespace. I.e., all of the following are valid expressions.

1-1    // 0
1 -1   // 0
1- 1   // 0
1 - 1  // 0
1- -1  // 2
1 - -1 // 2

6 + -(4)   // 2
6 + -( -4) // 10

And the following are invalid expressions

1 - - 1    // Invalid
1- - 1     // Invalid
6 + - (4)  // Invalid
6 + -(- 4) // Invalid

Validation

You do not need to worry about validation - you will only receive valid mathematical expressions following the above rules.

NOTE: eval and exec are disallowed in your solution.


```python
def calc(expression):
    
    #removing white spaces
    expression = expression.replace(' ', '')
        
    
    #removing parentesis parentesis
    while expression.count('(') != 0:
        ini = 0
        fin = 0
        for index in range(len(expression)):
            if expression[index] == '(':
                ini = index
            elif expression[index] == ')':
                fin = index
                break

        res = calc(expression[ini+1:fin])
        expression = expression[0:ini] + str(res) + expression[fin+1:]
     
    
    #cleaning double plus and minus symbols 
    while True:
        expression = expression.replace('--', '+')
        expression = expression.replace('+-', '-')
        expression = expression.replace('-+', '-')
        expression = expression.replace('++', '+')
        if (expression.count('--') == 0) and (expression.count('++') == 0) and (expression.count('+-') == 0) and (expression.count('-+') == 0):
            break
    
    
    #list with the possible terms
    lst_terms = []
    
    #symbol between terms
    the_symbol = ''
    
    
    #try to split in + or - symbols
    flag_pm = False
    pm_symbols = [pos for pos, char in enumerate(expression) if char in ['+', '-']]
    for middle in pm_symbols:
        if middle == 0:
            continue
        if expression[middle-1] in ['*', '/']:
            continue
        temp1 = expression[:middle]
        temp2 = expression[middle+1:]
        the_symbol = expression[middle]
        if temp1.count('(') == temp1.count(')'):
            flag_pm = True
            break
    
    if flag_pm:
        lst_terms.append(temp1)
        
        if the_symbol == '-':
            temp2 = temp2.replace('/-', 'Z')
            temp2 = temp2.replace('*-', 'Y')
            temp2 = temp2.replace('-', 'X')
            temp2 = temp2.replace('+', '-')
            temp2 = temp2.replace('X', '+')
            temp2 = temp2.replace('Z', '/-')
            temp2 = temp2.replace('Y', '*-')
        lst_terms.append(temp2)
       
    
    
    #if it is not possible to divide into + and -, try with *  and /
    flag_td = False
    if flag_pm == False:
        pm_symbols = [pos for pos, char in enumerate(expression) if char in ['*', '/']]
        for middle in reversed(pm_symbols):
            if middle == 0:
                continue
            temp1 = expression[:middle]
            temp2 = expression[middle+1:]
            the_symbol = expression[middle]
            if temp1.count('(') == temp1.count(')'):
                flag_td = True
                break
    
    if flag_td:
        lst_terms.append(temp1)
        lst_terms.append(temp2)
        
    
    #there is only one term (number) without symbols
    if flag_pm == False and flag_td == False:
        lst_terms.append(expression)
    
    
    if len(lst_terms) == 1:
        return float(lst_terms[0])
    else:
        L = calc(lst_terms[0])
        R = calc(lst_terms[1])
    
    if the_symbol == '+':
        return L + R
    if the_symbol == '-':
        return L - R
    if the_symbol == '*':
        return L * R
    if the_symbol == '/':
        return L / R if R != 0 else 0
    
    return None
```
